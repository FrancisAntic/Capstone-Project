####**HEARTRATE**

**Analysing Heart Rate of a user**

Here value column shows users heartbeat in BeatsPerMinute(BPM)

```python
heartrate.head()
```

```python
heartrate.drop(['Catg Day'], axis=1)
```

**Extracting Hour from Time column for analysing heart beat of a user at a particular time**

```python
heartrate['Time Am/Pm'] = heartrate['Time'].dt.hour
```

```python
heartrate['Time Category'].value_counts()
```

Categorising Time in Mid-night, Morning, Evening, Night

```python
#time division
bins = [0,6,12,18,23]
group = ["Mid-night","Morning","Evening","Night"]
heartrate['Time Category']=pd.cut(heartrate['Time Am/Pm'],bins,labels=group)
```

Categorising Heart Beat in Less, Good And High

```python
#heartrate['Time Am/Pm'].value_counts()
bins1 = [30,76,140,210]
group1=["Less","Good","High"]
heartrate['Catg']=pd.cut(heartrate['Value'],bins1,labels=group1)
```

```python
uh=pd.crosstab(heartrate['Id'],heartrate['Catg'])
uh.plot(kind='bar',title='User HeartRate',ylabel='Heartrate')
```

The above chart show that most of the users have less heart rate That is between 30 - 76 beats per minute

```python
vt=pd.crosstab(heartrate['Time Category'],heartrate['Catg'])
vt.plot(kind='bar',title="Heart Rate WRT Time Category",ylabel='Heartrate')
```

The Above chart shows that users have a good heart rate during the evening time
and

 morning time their heart rate is less

```python
heartrate['Catg'].value_counts().plot(kind='pie')
```

The Counts pie chart show that Users with more **Less heart rate than Good heart rate**
